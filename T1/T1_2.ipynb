```python
import pandas as pd
import numpy as np

sheet1 = pd.read_csv("../sheet0_num.csv").drop(["ornament", "color"],axis=1)
sheet2 = pd.read_csv("../sheet1_1.csv").drop(["weathered"],axis=1)
data=pd.merge(sheet1,sheet2,on='id')
```

```python
data=data.fillna(0).drop(["id"],axis=1)
data
```

```python
not_wind_a=data[(data['type']==0)&(data['degree']==0)]
wind_a=data[(data['type']==0)&(data['degree']==1)]
not_wind_b=data[(data['type']==1)&(data['degree']==0)]
wind_b=data[(data['type']==1)&(data['degree']==1)]
```

```python
def solve_system(x_var,x_mean,y_var,y_mean):
    """_summary_

    Args:
        x_var (_type_): 风化的方差
        x_mean (_type_): 风化的期望
        y_var (_type_): 未风化的方差
        y_mean (_type_): 未风化的期望
    """
    
    a=np.sqrt(y_var/x_var)
    b=y_mean-a*x_mean
    return a,b
```

```python
def convert_to_latex_style_table(not_wind, wind):
    name_list=['二氧化硅(SiO2)',
    '氧化钠(Na2O)',
    '氧化钾(K2O)',
    '氧化钙(CaO)',
    '氧化镁(MgO)',
    '氧化铝(Al2O3)',
    '氧化铁(Fe2O3)',
    '氧化铜(CuO)',
    '氧化铅(PbO)',
    '氧化钡(BaO)',
    '五氧化二磷(P2O5)',
    '氧化锶(SrO)',
    '氧化锡(SnO2)',
    '二氧化硫(SO2)']
    not_wind_var=not_wind.var()[2:]
    not_wind_mean = not_wind.mean()[2:]
    
    wind_var = wind.var()[2:]
    wind_mean = wind.mean()[2:]
    
    for i in range(len(name_list)):
        print(name_list[i], "&", format(not_wind_mean[i],'.4f'), "&", format(not_wind_var[i], '.4f'), "&", format(wind_mean[i], '.4f'), "&", format(wind_var[i], '.4f'), "//")
```

```python
def convert_to_latex_style_equ(a,b):
    a=a[2:]
    b=b[2:]
    name_list = ['二氧化硅(SiO2)',
                 '氧化钠(Na2O)',
                 '氧化钾(K2O)',
                 '氧化钙(CaO)',
                 '氧化镁(MgO)',
                 '氧化铝(Al2O3)',
                 '氧化铁(Fe2O3)',
                 '氧化铜(CuO)',
                 '氧化铅(PbO)',
                 '氧化钡(BaO)',
                 '五氧化二磷(P2O5)',
                 '氧化锶(SrO)',
                 '氧化锡(SnO2)',
                 '二氧化硫(SO2)']
    for i in range(len(a)):
        if b[i]>=0:
            print(name_list[i],"&","$y=%.4fx+%.4f$"%(a[i],b[i]),"//")
        else:
            print(name_list[i],"&","$y=%.4fx%.4f$"%(a[i],b[i]),"//")
```

```python
a_a,a_b=solve_system(wind_a.var(),wind_a.mean(),not_wind_a.var(),not_wind_a.mean())
b_a,b_b=solve_system(wind_b.var(),wind_b.mean(),not_wind_b.var(),not_wind_b.mean())
```

```python
convert_to_latex_style_equ(a_a,a_b)
convert_to_latex_style_equ(b_a,b_b)
```

```python
convert_to_latex_style_table(not_wind_a,wind_a)

convert_to_latex_style_table(not_wind_b,wind_b)
```
