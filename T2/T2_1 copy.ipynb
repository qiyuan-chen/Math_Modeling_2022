```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

all_variables = ['degree', 'sio', 'nao', 'ko', 'cao', 'mgo', 'alo', 'feo', 'cuo', 'pbo', 'bao', 'po', 'sro', 'sno', 'so']

composition_dict = {
    "degree": "é£ŽåŒ–ç¨‹åº¦",
    "sio": "äºŒæ°§åŒ–ç¡…(SiO2)",
    "nao": "æ°§åŒ–é’ (Na2O)"  ,
    "ko" : "æ°§åŒ–é’¾(K2O)"   ,
    "cao": "æ°§åŒ–é’™(CaO)"   , 
    "mgo": "æ°§åŒ–é•(MgO)"   ,
    "alo": "æ°§åŒ–é“(Al2O3)" ,
    "feo": "æ°§åŒ–é“(Fe2O3)" ,
    "cuo": "æ°§åŒ–é“œ(CuO)"   ,
    "pbo": "æ°§åŒ–é“…(PbO)"   ,
    "bao": "æ°§åŒ–é’¡(BaO)"   ,
    "po" : "äº”æ°§åŒ–äºŒç£·(P2O5)",
    "sro": "æ°§åŒ–é”¶(SrO)"   ,
    "sno": "æ°§åŒ–é”¡(SnO2)"  ,
    "so" : "äºŒæ°§åŒ–ç¡«(SO2)" ,
}

plt.figure(dpi=300) # è®¾ç½®åˆ†è¾¨çŽ‡
data = pd.read_csv("../std_data.csv").drop(columns=['id','type'])
data.columns = data.columns.map(composition_dict)
sns.heatmap(data.corr(), cmap='YlGnBu')
# plt.savefig("./corr_heatmap.png")
```

```python
# åˆ©ç”¨ ç›¸å…³æ€§ç³»æ•°çŸ©é˜µ è¿›è¡Œ å˜é‡é€‰æ‹©
# è®¾ç½® threshold ï¼Œå½“ä¸¤ä¸ªå˜é‡çš„ç›¸å…³ç³»æ•°çš„ç»å¯¹å€¼ å¤§äºŽ threshold æ—¶ï¼Œè®¤ä¸ºä¸¤ä¸ªå˜é‡ç›¸å…³ï¼Œéœ€è¦å…¶ä¸­åˆ æŽ‰ä¸€ä¸ªå˜é‡

THRESHOLD = 0.6

data = pd.read_csv("../std_data.csv").drop(columns=['id','type'])
corr = data.corr()
corr = corr.apply(lambda x: np.abs(x) > THRESHOLD)

relevant_variable_pairs = []
chosen_variables = []
vis = {}

def dfs(u):
    for v in all_variables:
        if (v == u) or (not corr[u][v]) or (vis.get(v) is not None):
            continue
        relevant_variable_pairs.append((u, v))
        vis[v] = True
        dfs(v)

for u in all_variables:
    if vis.get(u) is None:
        vis[u] = True
        chosen_variables.append(u)
        dfs(u)

print(f'relevant_variable_pairs: {relevant_variable_pairs}')
print(f'chosen_variables: {chosen_variables}')
```

```python
# ç¼ºå¤±å€¼å¤šçš„å±žæ€§åˆ æŽ‰
# å½“æŸä¸ªå±žæ€§çš„ç¼ºå¤±å€¼æ‰€å çš„æ¯”ä¾‹è¾¾åˆ° THRESHOLD æ—¶ï¼Œè®¤ä¸ºåº”è¯¥åˆ æŽ‰è¯¥å±žæ€§

all_variables = chosen_variables[:]
chosen_variables = []
dropped_variables = []

data = data[all_variables]

THRESHOLD = 0.75

for variable in all_variables:
    print(f'variable: {variable}, {len(data[data[variable] < 1e-3]) / len(data)}')
    if len(data[data[variable] < 1e-3]) / len(data) > THRESHOLD:
        dropped_variables.append(variable)
    else:
        chosen_variables.append(variable)

print(f'dropped_variables: {dropped_variables}')
print(f'chosen_variables: {chosen_variables}')
```

```python
# æ–¹å·®è¿‡å°çš„å±žæ€§åˆ æŽ‰

all_variables = chosen_variables[:]
chosen_variables = []
dropped_variables = []

data = data[all_variables]

THRESHOLD = 1.0

for variable in all_variables:
    print(f'variable {variable}, std = {np.std(data[variable])}')
    if np.std(data[variable]) < THRESHOLD:
        dropped_variables.append(variable)
    else:
        chosen_variables.append(variable)

print(f'dropped_variables: {dropped_variables}')
print(f'chosen_variables: {chosen_variables}')
```

```python
data = data[chosen_variables]
data = pd.concat([data, pd.read_csv('../std_data.csv')[['type']]], axis=1)
data.shape
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

pca = PCA(n_components=5)
data_train = data.drop(["type"], axis=1)
color = data["type"]
data_after_pca = pca.fit_transform(data_train)
print(pca.explained_variance_ratio_)
data_after_pca = pd.DataFrame(data_after_pca)
print(data_after_pca)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
colors = []
colors_ = []
mask = []
mask_ = []
for i in range(len(color)):
    if color[i] == 1:
        # é«˜é’¾
        colors.append('#8ECFC9')
        mask.append(True)
        mask_.append(False)
    else:
        # é’¡é“…
        colors_.append('#FA7F6F')
        mask_.append(True)
        mask.append(False)
%matplotlib inline
plt.figure(figsize=[5, 5])
sns.scatterplot(x=(np.array(data_after_pca[0])[mask]),
                y=(np.array(data_after_pca[1])[mask]), color='#8ECFC9', label='é«˜é’¾')
sns.scatterplot(x=(np.array(data_after_pca[0])[mask_]),
                y=(np.array(data_after_pca[1])[mask_]), color='#FA7F6F', label='é’¡é“…')
plt.xlabel("Principal Components 1")
plt.ylabel("Principal Components 2")
plt.legend()
plt.savefig("pca_fig.png")
```

```python
k1_spss = pca.components_.T
weight = (np.dot(k1_spss, pca.explained_variance_ratio_)) / \
    np.sum(pca.explained_variance_ratio_)
print('weight:', weight)
#
weighted_weight = weight/np.sum(weight)
print('weighted_weight:', weighted_weight)

print(np.abs(weight).argsort())
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


X=data.drop(columns=['type'])
Y=data['type']

good_vari = ['nao', 'cao', 'ko', 'cuo']

ALPHA = 1e-1

while True:
    for seed in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=seed)
        lasso = Lasso(alpha=ALPHA)
        lasso.fit(X_train, y_train)
        y_pred_train=lasso.predict(X_train)
        y_pred_test=lasso.predict(X_test)
        # print(r2_score(y_train,y_pred_train))
        # print(r2_score(y_test,y_pred_test))
        coef = pd.Series(lasso.coef_, index=data.drop(columns=['type']).columns)
        rst = coef.abs().sort_values(ascending = False)[:4].to_dict()
        cnt = 0
        for x in good_vari:
            if rst.get(x) is not None:
                cnt+=1
        if cnt >= 4 and (r2_score(y_test,y_pred_test) > 0.7):
            print(ALPHA, seed)
            break
    ALPHA /= 2
    if ALPHA < 1e-2:
        break
```
