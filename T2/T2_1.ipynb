```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

all_variables = ['degree', 'sio', 'nao', 'ko', 'cao', 'mgo', 'alo', 'feo', 'cuo', 'pbo', 'bao', 'po', 'sro', 'sno', 'so']

composition_dict = {
    "degree": "风化程度",
    "sio": "二氧化硅(SiO2)",
    "nao": "氧化钠(Na2O)"  ,
    "ko" : "氧化钾(K2O)"   ,
    "cao": "氧化钙(CaO)"   , 
    "mgo": "氧化镁(MgO)"   ,
    "alo": "氧化铝(Al2O3)" ,
    "feo": "氧化铁(Fe2O3)" ,
    "cuo": "氧化铜(CuO)"   ,
    "pbo": "氧化铅(PbO)"   ,
    "bao": "氧化钡(BaO)"   ,
    "po" : "五氧化二磷(P2O5)",
    "sro": "氧化锶(SrO)"   ,
    "sno": "氧化锡(SnO2)"  ,
    "so" : "二氧化硫(SO2)" ,
}

plt.figure(dpi=400) # 设置分辨率
data = pd.read_csv("../std_data.csv").drop(columns=['id','type'])
data.columns = data.columns.map(composition_dict)
sns.heatmap(data.corr(), cmap='YlGnBu')
plt.savefig("./corr_heatmap.png")
```

```python
# 利用 相关性系数矩阵 进行 变量选择
# 设置 threshold ，当两个变量的相关系数的绝对值 大于 threshold 时，认为两个变量相关，需要其中删掉一个变量

THRESHOLD = 0.6

data = pd.read_csv("../std_data.csv").drop(columns=['id','type'])
corr = data.corr()
corr = corr.apply(lambda x: np.abs(x) > THRESHOLD)

relevant_variable_pairs = []
chosen_variables = []
vis = {}

def dfs(u):
    for v in all_variables:
        if (v == u) or (not corr[u][v]) or (vis.get(v) is not None):
            continue
        relevant_variable_pairs.append((u, v))
        vis[v] = True
        dfs(v)

for u in all_variables:
    if vis.get(u) is None:
        vis[u] = True
        chosen_variables.append(u)
        dfs(u)

print(f'relevant_variable_pairs: {relevant_variable_pairs}')
print(f'chosen_variables: {chosen_variables}')
```

```python
# 缺失值多的属性删掉
# 当某个属性的缺失值所占的比例达到 THRESHOLD 时，认为应该删掉该属性

all_variables = chosen_variables[:]
chosen_variables = []
dropped_variables = []

data = data[all_variables]

THRESHOLD = 0.75

for variable in all_variables:
    print(f'variable: {variable}, {len(data[data[variable] < 1e-3]) / len(data)}')
    if len(data[data[variable] < 1e-3]) / len(data) > THRESHOLD:
        dropped_variables.append(variable)
    else:
        chosen_variables.append(variable)

print(f'dropped_variables: {dropped_variables}')
print(f'chosen_variables: {chosen_variables}')
```

```python
# 方差过小的属性删掉

all_variables = chosen_variables[:]
chosen_variables = []
dropped_variables = []

data = data[all_variables]

THRESHOLD = 1.0

for variable in all_variables:
    print(f'variable {variable}, std = {np.std(data[variable])}')
    if np.std(data[variable]) < THRESHOLD:
        dropped_variables.append(variable)
    else:
        chosen_variables.append(variable)

print(f'dropped_variables: {dropped_variables}')
print(f'chosen_variables: {chosen_variables}')
```

```python
data = data[chosen_variables]
data = pd.concat([data, pd.read_csv('../std_data.csv')[['type']]], axis=1)
data
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


X=data.drop(columns=['type'])
Y=data['type']

good_vari = ['nao', 'cao', 'ko', 'cuo']

ALPHA = 1e-1

while True:
    for seed in range(10):
        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=seed)
        lasso = Lasso(alpha=ALPHA)
        lasso.fit(X_train, y_train)
        y_pred_train=lasso.predict(X_train)
        y_pred_test=lasso.predict(X_test)
        # print(r2_score(y_train,y_pred_train))
        # print(r2_score(y_test,y_pred_test))
        coef = pd.Series(lasso.coef_, index=data.drop(columns=['type']).columns)
        rst = coef.abs().sort_values(ascending = False)[:4].to_dict()
        cnt = 0
        for x in good_vari:
            if rst.get(x) is not None:
                cnt+=1
        if cnt >= 4 and (r2_score(y_test,y_pred_test) > 0.7):
            print(ALPHA, seed)
            break
    ALPHA /= 2
    if ALPHA < 1e-2:
        break
```
