```python
import pandas as pd
import numpy as np

sheet1 = pd.read_csv("../sheet0_num.csv").drop(["ornament", "color"],axis=1)
sheet2 = pd.read_csv("../sheet1_1.csv").fillna(0)
```

```python
sheet2 = pd.concat([
    sheet2[['id',]], 
    sheet2.drop(columns=['id', 'weathered']).apply(
        lambda x: ((x-np.min(x))/(np.max(x)-np.min(x)) * 100)
    )
], axis=1)
sheet2
```

```python
data=pd.merge(sheet1, sheet2,on="id")
data
```

```python
data.to_csv('../std_data.csv', index=False)
```
